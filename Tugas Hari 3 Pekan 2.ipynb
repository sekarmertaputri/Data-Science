<h2>Soal 2: Data Restructuring</h2>

Jelaskan itu pivoting dan melting data

Pivoting adalah proses untuk merubah data yang awalnya berada dalam baris menjadi kolom. Berfungsi untuk mengubak data yang dari format panjang (long format) menjadi format lebar (wide format).
Melting adalah proses sebaliknya dari pivoting yaitu merubak data yang awalnya berada di kolom menjadi baris.

<h2>Soal 3: Reshaping Data 1</h2>

```python
import pandas as pd
import random

data = {
    'Tipe Ujian' : ['UTS1', 'UAS1', 'UTS2', 'UAS2']*3,
    'Mata Pelajaran' : ['Fisika']*4 + ['Matematika']*4+['Statistika']*4,
    'Rata2 Waktu Belajar (menit)' : [random.choice(range(200, 800)) for i in range(12)],
    'Rata2 Nilai Ujian' : [random.choice(range(40, 100)) for i in range(12)] 
}

df = pd.DataFrame(data)
df
```

<h2>Soal 4: Reshaping Data 2</h2>

```python
import pandas as pd

data_1 = {
    'Hari' : ['Sabtu', 'Minggu'],
    'Kebun Binatang' : [271, 399],
    'Dufan' : [501, 700],
    'Disney' : [1000, 1001],
    'Bali' : [900, 803]
}
df_1 = pd.DataFrame(data_1).set_index('Hari')
df_1
```

diatas adalah data tentang jumlah pengunjung di berbagai tempat hiburan, ubahlah struktur data tersebut sehingga sesuai dengan hasil yang diharapkan.

```python
#code here
melt_df_1 = pd.melt(df_1.reset_index(), id_vars=['Hari'], var_name='Tempat Hiburan', value_name='Pengunjung')
melt_df_1
```
