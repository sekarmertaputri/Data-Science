<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,8), facecolor='#808080', sharex=True, sharey=True)

ax[0].plot(x, tan, color='#0000FF', label='tanX')
ax[0].set_title('Grafik tanx')
ax[1].plot(x, sin, color='#FF0000', label='sinX')
ax[1].set_title('Grafik sinx')
ax[2].plot(x, cos, color='#008000', label='cosX')
ax[2].set_title('Grafik cosx')

#membuat lokasi keterangan label
fig.legend(loc='upper center')

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here

fig, ax = plt.subplots(figsize=(12,8), facecolor='#808080')

ax1 = fig.add_axes([0.1, 0.1, 1, 1]) #fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, color='#4169E1')
ax1.set_title('Grafik CosX', y=1.2)

ax2 = fig.add_axes([0, 0, 0.3, 0.3])
ax2.plot(x, y2, color='#FF0000')
ax2.plot(x, y, color='#0000FF')
ax2.set_title('Grafik CosX^2')

ax3 = fig.add_axes([1, 0, 0.3, 0.3])
ax3.plot(x, y3, color='#228B22')
ax3.plot(x, y, color='#0000FF')
ax3.set_title('Grafik CosX^3')

ax4 = fig.add_axes([1, 1, 0.3, 0.3])
ax4.plot(x, y4, color='#000000')
ax4.plot(x, y, color='#0000FF')
ax4.set_title('Grafik CosX^4')

ax5 = fig.add_axes([0, 1, 0.3, 0.3])
ax5.plot(x, y5, color='#FFD700')
ax5.plot(x, y, color='#0000FF')
ax5.set_title('Grafik CosX^5')

plt.show()
```
