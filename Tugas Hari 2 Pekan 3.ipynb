Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
#code here

# membuat component figure dan axis
fig, ax = plt.subplots(figsize=(11, 9), facecolor='#FFFFFF')

#memberikan data ke dalam axis
ax.plot(x, cos, linestyle='solid', color='#FF0000', label='Cosx')
ax.plot(x, cos3, linestyle='dashed', color='#00FF00', label='3 Cosx')
ax.plot(x, sin, linestyle='dashdot', color='#0000FF', label='Sin(x-90)')
ax.plot(x, sin2, linestyle='', marker='x', color='#FFD700', label='2 Sinx')

# mengubah axis label dan title
ax.set_xlabel('X-Axis Label', fontsize='small')
ax.set_ylabel('Y-Axis Label', fontsize='small')
ax.set_title('Grafik Cos Sin', fontsize='medium')
ax.legend(loc='center', fontsize='small')
ax.tick_params(labelsize='small')

plt.show()
```
