**Soal 3. Filtering**

Tampilkan data perubahan iklim pada 1 bulan terakhir di tahun 2016 dan jawablah pertanyaan berikut:

Berapakah rata-rata temperatur tertinggi?

```python
#code here
#note! -> .last('M') m disini yaitu month/bulan karena data yg dicari u/ 1 bulan
#tapi karna update versi jadi M diubah ke ME

bulan_terakhir_2016 = df.loc['2016'].last('ME')
bulan_terakhir_2016
```

```python
#code here

df_sorted = bulan_terakhir_2016.sort_values('meantemp', ascending=False)
print('\nRata-Rata Temperatur (MeanTemp) Tertinggi yaitu : \n')
df_sorted.head(1)
```

**Soal 4. Time grouping**

 

*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
#code here

agregasi_bulanan_2016 = df.loc['2016'].resample('ME').agg({
    'meantemp' : 'max', 'humidity' : 'max', 'wind_speed' : 'max', 'meanpressure' : 'max'
})

agregasi_bulanan_2016
```

```python
#code here

df_sorted_agregasi_meantemp = agregasi_bulanan_2016.sort_values('meantemp',  ascending=False)
print('\nRata-Rata Temperatur (MeanTemp) Tertinggi yaitu di bulan : \n')
df_sorted_agregasi_meantemp.head(1)
```

```python
#code here

df_sorted_agregasi_humidity = agregasi_bulanan_2016.sort_values('humidity',  ascending=False)
print('\nhumidity Tertinggi yaitu di bulan : \n')
df_sorted_agregasi_humidity.head(1)
```

```python
#code here

df_sorted_agregasi_wind_speed = agregasi_bulanan_2016.sort_values('wind_speed',  ascending=False)
print('\nwind_speed Tertinggi yaitu di bulan : \n')
df_sorted_agregasi_wind_speed.head(1)
```

```python
#code here

df_sorted_agregasi_meanpressure = agregasi_bulanan_2016.sort_values('meanpressure',  ascending=False)
print('\nmeanpressure Tertinggi yaitu di bulan : \n')
df_sorted_agregasi_meanpressure.head(1)
```
