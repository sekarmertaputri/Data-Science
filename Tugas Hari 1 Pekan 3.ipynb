<h2>Soal 1: Pemahaman Component Matplotlib</h2>

- Jelaskan apa itu Figure
- Jelaskan apa itu Axis

Figure adalah window atau page atau halaman dalam objek visual. Figure bisa diartikan sebagai wadah untuk menggambah seperti halnya saat menggambar maka figure yang digunakan adalah kertas gambar.

Axes (axis) adalah suatu area di dalam figure dimana data akan di plot, kedalam Figure kita menambahkan Axes. Dalam Axes juga terdapat berbagai macam komponen lagi seperti x-axis, y-axis, dan lain sebagainya. 

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# membuat Figure dan Axis dengan Data Kosong

# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menampilkan objek visual
plt.show()
```

```python
#eksplisit

# membuat objek figure
fig_1 = plt.figure()

# membuat objek axes
ax_1 = fig_1.add_subplot()

# menambahkan dalam ke objek axes
ax_1.plot(x, y)

# menampilkan data
plt.show()
```

```python
#implicit

# membuat objek visual
plt.plot(x, y)

# menampilkan objek visual
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas (apa saja) dan tampilkan kembali gambar tersebut.

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
#print(plt.style.available) cara cek package apa saja yg tersedia

plt.style.use('seaborn-v0_8')

fig_2, ax_2 = plt.subplots()
ax_2.plot(x, y)
ax_2.plot(x, y2)
ax_2.plot(x, y3)
plt.show()
```
